```python
# Basic data manipulation libraries
import pandas as pd
import numpy as np

# Visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Statistical testing
from scipy import stats

# For parsing JSON-like strings in TMDB dataset
import ast

# Set visualization style for cleaner plots
sns.set(style="whitegrid")

# Make plots appear inside notebook
%matplotlib inline
```

```python
# Load datasets
movies = pd.read_csv(r"C:\Users\senak\Desktop\dsa210 project\data\raw\tmdb_5000_movies.csv")
credits = pd.read_csv(r"C:\Users\senak\Desktop\dsa210 project\data\raw\tmdb_5000_credits.csv")
bechdel = pd.read_csv(r"C:\Users\senak\Desktop\dsa210 project\data\raw\Bechdel_IMDB_Merge0524.csv")

# Check dataset shapes
print("Movies:", movies.shape)
print("Credits:", credits.shape)
print("Bechdel:", bechdel.shape)
```

```python
# Merge movies and credits datasets
# Both datasets share movie id

df = movies.merge(
    credits,
    left_on="id",
    right_on="movie_id"
)

# Check result
df.head()
```

```python
# Function to calculate female ratio from cast column

def get_female_ratio(cast):

    # Convert string to list
    cast = ast.literal_eval(cast)

    female_count = 0
    total_count = 0

    for person in cast:
        if person['gender'] == 1:
            female_count += 1
        total_count += 1

    # Avoid division by zero
    if total_count == 0:
        return 0

    return female_count / total_count


# Apply function
df['female_ratio'] = df['cast'].apply(get_female_ratio)
```

```python
# Function to detect female director

def female_director(crew):

    crew = ast.literal_eval(crew)

    for person in crew:
        if person['job'] == "Director" and person['gender'] == 1:
            return 1

    return 0


# Apply
df['female_director'] = df['crew'].apply(female_director)
```

```python
# Convert release date to datetime
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')

# Extract year
df['year'] = df['release_date'].dt.year
```

```python
# After merging movies and credits datasets,
# pandas created title_x and title_y columns.
# I will keep the one coming from the movies dataset.

df = df.rename(columns={'title_x': 'title'})
df = df.drop(columns=['title_y'])

# quick check
df[['title', 'year']].head()
```

```python
# Clean title columns to reduce matching issues

df['title'] = df['title'].astype(str).str.strip()
bechdel['title'] = bechdel['title'].astype(str).str.strip()

# Convert years to numeric
df['year'] = pd.to_numeric(df['year'], errors='coerce')
bechdel['year'] = pd.to_numeric(bechdel['year'], errors='coerce')
```

```python
# Merge TMDB data with Bechdel dataset

final_df = df.merge(
    bechdel,
    on=['title', 'year'],
    how='left'
)

print("Shape after merge:", final_df.shape)
final_df.head()
```

```python
# Check how many movies matched

matched = final_df['bechdelRating'].notna().sum()
total = len(final_df)

print("Matched movies:", matched)
print("Total movies:", total)
print("Match rate:", matched / total)
```

```python
# Keeping only the variables that will be useful for my analysis

final_df = final_df[[
    'title',
    'year',
    'budget',
    'revenue',
    'vote_average',
    'vote_count',
    'popularity',
    'female_ratio',
    'female_director',
    'bechdelRating',
    'imdbAverageRating',
    'numVotes',
    'runtimeMinutes',
    'genre1',
    'genre2',
    'genre3'
]]

final_df.head()
```

```python
# check missing values

final_df.isna().sum().sort_values(ascending=False)
```

```python
# remove unrealistic budgets

final_df = final_df[final_df['budget'] > 0]
final_df = final_df[final_df['revenue'] > 0]

print(final_df.shape)
```
